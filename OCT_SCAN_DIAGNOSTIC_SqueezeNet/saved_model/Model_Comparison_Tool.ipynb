```python
import pandas as pd
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
import matplotlib.pyplot as plt
```

```python
def Plot_Comparison(X,X1,legend_Text,Title_Text):
    """
    Plot curves comparison
    Input Data : X,X1,legend_Text,Title_Text -> Input Dataframe, Input Datafram1, legend text, Title text
    """
    legend_acc = [legend_Text[0] + ' acc',legend_Text[0] + ' val_acc',legend_Text[1] + ' acc',legend_Text[1] + ' val_acc']
    fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15, 5))

    X['acc'].plot(style='r',ax=axes[0])
    X['val_acc'].plot(style='r--',ax=axes[0])
    X1['acc'].plot(style='g',ax=axes[0])
    X1['val_acc'].plot(style='g--',ax=axes[0])
    # axes[0].set_title('X1 val acc')
    axes[0].xaxis.set_minor_locator(AutoMinorLocator())
    axes[0].yaxis.set_minor_locator(AutoMinorLocator())
    axes[0].xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
    axes[0].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    axes[0].set_ylabel('Accuracy', fontweight='bold')
    axes[0].set_xlabel('Epoch', fontweight='bold')
    axes[0].set_ylim(ymin=0)
    axes[0].set_xlim(xmin=0)
    axes[0].legend(legend_acc, loc='best')
    axes[0].grid()

    legend_loss = [legend_Text[0] + ' loss',legend_Text[0] + ' val_loss',legend_Text[1] + ' loss',legend_Text[1] + ' val_loss']
    X['loss'].plot(style='r',ax=axes[1])
    X['val_loss'].plot(style='r--',ax=axes[1])
    X1['loss'].plot(style='g',ax=axes[1])
    X1['val_loss'].plot(style='g--',ax=axes[1])
    # axes[0].set_title('X1 val acc')
    axes[1].xaxis.set_minor_locator(AutoMinorLocator())
    axes[1].yaxis.set_minor_locator(AutoMinorLocator())
    axes[1].xaxis.set_major_formatter(FormatStrFormatter('%.0f'))
    axes[1].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    axes[1].set_ylabel('Loss', fontweight='bold')
    axes[1].set_xlabel('Epoch', fontweight='bold')
    axes[1].set_ylim(ymin=0)
    axes[1].set_xlim(xmin=0)
    axes[1].legend(legend_loss, loc='best')
    axes[1].grid()
    
    fig.suptitle(Title_Text)
#     fig.savefig('a.png', bbox_inches='tight', dpi=600)

    return
```

# Initial Kernel Sensitivity

```python
CsvFile='SqueezeNet_Model_From_Scratch.hdf5.csv'
CsvFile1='SqueezeNet_Model_From_Scratch_InitRandomWeight_ImageResize.hdf5.csv'
legend_text=['glorot_uniform_kernel + Image resize 227x227','he_normal_kernel + Image resize 350x350']
Title_Text='Kernel Sensitivity\nglorot_uniform_kernel Vs he_normal_kernel'
X=pd.read_csv(CsvFile)
X1=pd.read_csv(CsvFile1)

Plot_Comparison(X,X1,legend_text,Title_Text)
```
